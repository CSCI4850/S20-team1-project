```python
from keras.preprocessing import image
from keras.models import load_model
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_shape = (150,150,3)
classes={'busy': 0, 'free': 1}
model = load_model("moretransform.h5")
```

```python
import cv2
import numpy as np
from PIL import Image
def myFunc(image):
    image = np.array(image)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    img2 = np.zeros_like(image)
    img2[:,:,0] = gray
    img2[:,:,1] = gray
    img2[:,:,2] = gray
    return img2
def increase_brightness(img, value=30):
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    h, s, v = cv2.split(hsv)

    lim = 255 - value
    v[v > lim] = 255
    v[v <= lim] += value

    final_hsv = cv2.merge((h, s, v))
    img = cv2.cvtColor(final_hsv, cv2.COLOR_HSV2BGR)
    return img
```

# Parking lot 1

```python
img = image.load_img("./parking lot 1/parkinglot1.jpg")
img = image.img_to_array(img)
plt.imshow(img/255)
plt.show()

dirname = "./parking lot 1/images"
free = 0
busy = 0
for i in range(1, 76):
    filepath = os.path.join(dirname, str(i)+".png");
    img = image.load_img(filepath, target_size=image_shape)
    img = image.img_to_array(img)
    img = increase_brightness(img, -50)
#         plt.imshow(img/255)
#         plt.show()
    img = img.reshape((-1, 150, 150, 3))
    result = int(model.predict(img)[0][0])
    result = "free" if result else "busy"

    if (result == "free"):
        free += 1
#         print(filename + " free")
    else:
#         print(filename + " busy")
        busy += 1

print("Predicted:")
print("busy: ", busy)
print("free: ", free)

print("\nActual:")
print("busy: 45")
print("free: 30")

missed = (abs(busy-45)+(abs(free-30)))
print("Accuracy: ", 1-(missed/75))
```
